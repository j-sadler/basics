```python
#  how do I want to filter the data table?  make new variables with the values that you want.

# 'Portfolio [SPI]'
Portfolio = 'Enterprise'
# DT['Funding Status [SPI]']
FundingStatus = ['Approved','Approved (Non-Target)','Prior Year']
# DT['Funding Type [SPI]']
FundingType = 'BA'
```

```python
#below I can use unique to see the available values
DT['Funding Status [SPI]'].unique()
```

```python
DTFiltered = DT[(DT['Portfolio [SPI]'] == Portfolio) & (DT['Funding Status [SPI]'].isin(FundingStatus)) & (DT['Funding Type [SPI]'] == FundingType)]
```
